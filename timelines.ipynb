```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Define the timeline data
dates = [
    "2024-01-01", "2024-02-01", "2024-03-01", 
    "2024-04-01", "2024-05-01"
]
events = [
    "Project Kickoff", "Initial Research", "Prototype Development", 
    "User Testing", "Final Delivery"
]

# Convert string dates to datetime objects
dates = [datetime.strptime(date, "%Y-%m-%d") for date in dates]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(dates, [1] * len(dates), "|", color="darkblue", markersize=30)  # Timeline markers

# Add annotations for each event
for i, (date, event) in enumerate(zip(dates, events)):
    ax.text(date, 1.05, event, rotation=45, ha="right", fontsize=9)

# Formatting the timeline
ax.get_yaxis().set_visible(False)
ax.set_xlim([min(dates), max(dates)])
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

# Remove unnecessary chart elements
plt.tight_layout()

# Save as image
plt.savefig("timeline.png", dpi=300, bbox_inches="tight")
plt.show()
```
