```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate baseline data (training data)
np.random.seed(42)
baseline_data = np.random.normal(loc=0, scale=1, size=1000)

# Simulate incoming data with a slight shift (distribution shift)
shifted_data = np.random.normal(loc=1, scale=1, size=1000)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot baseline data distribution (Training data)
sns.kdeplot(baseline_data, label="Baseline Data (Training)", color='blue', fill=True)

# Plot incoming data distribution (New data in production)
sns.kdeplot(shifted_data, label="Shifted Data (Production)", color='red', fill=True)

# Add labels and title
plt.title('Visualizing Data Shift in MLOps', fontsize=14)
plt.xlabel('Feature Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

# Save the plot as a PNG file
plt.savefig('data_shift_distribution.png', format='png')
# Show plot
plt.show()
```
